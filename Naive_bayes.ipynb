```python
X =df.drop(['Class'], axis=1)
Y=df[['Class']]
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state=0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)

```

```python
#Importing the Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB

#Creating a Gaussian Naive Bayes instance:
nb = GaussianNB()

#Fitting the model to the training data:
nb.fit(X_train, Y_train)

#Making predictions on the training data:
predTrain = nb.predict(X_train)

#Importing the classification_report function to generate a classification report:
from sklearn.metrics import classification_report

#Printing the classification report for the training predictions:
print(classification_report(Y_train, predTrain))

#Importing the roc_auc_score function:
from sklearn.metrics import roc_auc_score as roc

#Calculating and printing the ROC AUC score for the training predictions:
print('ROC AUC Score: ',roc(Y_train, predTrain))
```
